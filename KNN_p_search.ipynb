```python
import pandas as pd
import numpy as np

# Define the categories for city and property_type
cities = ["Pune","satara","Anagar","Snagar","Katraj","swargare"] # 2000 rows with "Pune"
property_types = ["industrial", "commercial", "IT companies", "offices", "schools", "commercial"]

# Generate random data for price and building size
price = np.random.randint(50000, 2000000, size=2000)
building_size = np.random.randint(500, 3000, size=2000)

# Randomly choose property types for each row
property_type = np.random.choice(property_types, size=2000)
cities= np.random.choice(cities, size=2000)

# Create the DataFrame
dataset = pd.DataFrame({
    "city": cities,
    "property_type": property_type,
    "price": price,
    "building size": building_size
})

# Display the first few rows of the generated data
print(dataset.head())
```

```python
dataset
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Assuming you have a DataFrame named 'dataset'
X = dataset.copy()

# Define categorical and continuous columns
categorical_columns = ['city', 'property_type']
continuous_columns = ['price', 'building size']

# Create a preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X.columns.get_indexer(continuous_columns)),
        ('cat', OneHotEncoder(), X.columns.get_indexer(categorical_columns))
    ])

# Concatenate preprocessing and Nearest Neighbors into a single pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', NearestNeighbors(n_neighbors=5, metric='cosine'))
])

# Fit the model on the entire dataset
model.fit(X)

# Define a query data point for which you want to find similar entries
query_data = pd.DataFrame({
    'city': ['Pune'],
    'property_type': ['commercial'],
    'price': [120000],
    'building size': [1500]
})

# Transform the query data using the preprocessor
query_data_transformed = model.named_steps['preprocessor'].transform(query_data)

# Find the indices of the nearest neighbors
distances, indices = model.named_steps['model'].kneighbors(query_data_transformed)

# Display the indices of similar data points
print("Indices of similar data points:", indices)
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Assuming you have a DataFrame named 'dataset'
X = dataset.copy()

# Define categorical and continuous columns
categorical_columns = ['city', 'property_type']
continuous_columns = ['price', 'building size']

# Create a preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X.columns.get_indexer(continuous_columns)),
        ('cat', OneHotEncoder(), X.columns.get_indexer(categorical_columns))
    ])

# Concatenate preprocessing and Nearest Neighbors into a single pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', NearestNeighbors(n_neighbors=50, metric='cosine'))
])

# Fit the model on the entire dataset
model.fit(X)

# Define a query data point for which you want to find similar entries
query_data = pd.DataFrame({
    'city': ['satara'],
    'property_type': ['commercial'],
    'price': [120000],
    'building size': [1500]
})

# Transform the query data using the preprocessor
query_data_transformed = model.named_steps['preprocessor'].transform(query_data)

# Find the indices of the nearest neighbors
distances, indices = model.named_steps['model'].kneighbors(query_data_transformed)

# Extract rows from the original dataset based on the indices
matching_data = X.iloc[indices.flatten()]

# Display the DataFrame of matching data
print("Matching Data:")
print(matching_data)
```
